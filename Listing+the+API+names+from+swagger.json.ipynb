### The code takes a Swagger documentation in .sjon format and lists out the API names

```python
import json
from pprint import pprint
```

```python
import json
from pprint import pprint

data = json.load(open('swagger.json'))

pprint(data)
```

```python
operationIds = []
for key1 in data["paths"]:
    for key2 in data["paths"][key1]:
        for key3 in data["paths"][key1][key2]:
            if(key3 == "operationId"):
                operationIds.append(data["paths"][key1][key2].get(key3))   
                
print("List of Services: \n", operationIds)
```

```python
import pandas as pd
df = pd.DataFrame(operationIds,columns=['API Name'])
df.head()
```

```python
writer = pd.ExcelWriter('API List.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
